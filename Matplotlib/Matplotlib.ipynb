```python
# hm base ke line ko bhi given color me change kr skte hai
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="python")

plt.legend()
plt.show()
```

```python
# hm apne base ke acc line ke color ko bhi change kr skte hai

import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="python") # use_line_collection=False, currnetly outdated

plt.legend()
plt.show()
```

```python
# we can also change the orientation of this
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x, y,
         linefmt=":",        # dotted stems
         markerfmt="ro",     # red circle markers
         basefmt="g",        # baseline format (green)
         bottom=0,
         label="python",
         orientation="horizontal")   # fixed: no trailing space

plt.legend()
plt.show()
```

# Fill_Between

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,color="r")

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# agar hm line ke niche ki value ko fill karna chahte hai then use
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x,y)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# agar hme given value ke hisab se area fill karana hai then use
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x=[2,4],y1=2,y2=4) # positional argument between the diff area(1st,2nd)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# we can also change the color inside this
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g")

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# hm where argument ka use karke given area ko fill kr skte hai, without giving the value of y(where will show no error)
# but we must have to pass the where inside the list so use numpy to pass this inside the list of array

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.5) # alpha is used to make this area light.

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
