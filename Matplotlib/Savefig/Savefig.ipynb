# Savefig

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("line")

plt.show()
```

```python
# dpi(dot per inche) parameter ye zoom karne ke baad bhi line ki quality ko maintain rkhata hai
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("line1",dpi=2000)

plt.show()
```

```python
# hm graph ka facecolor bhi change kr skte hai -> means padding ki boundary ko bhi colorful bna skte hai

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("line2",dpi=2000,facecolor="g")

plt.show()
```

```python
# hm apne file ka formate bhi change kr skte hai like (pdf,jpg etc.)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("line3.pdf",dpi=2000,facecolor="g")

plt.show()
```

```python
# hm apne file ko jpg formate me bhi kr skte hai
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("line4.jpg",dpi=2000,facecolor="g")

plt.show()
```

```python
# hm apne graph ko transparent bhi bna skte hai(means isme background color bhi show hoga)
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("line5",dpi=2000,facecolor="g",
           transparent=True)

plt.show()
```

```python
# hm apne box ke inches ko bhi change kr skte hai
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y,color="r")

plt.savefig("line6",dpi=2000,facecolor="g",
           transparent=False,bbox_inches="tight")

plt.show()
```

# axis functions in matplotlib

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.show()
```

```python
# agar hme axis ka data change karna hai then use xtick() and yticks() function
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x)
plt.yticks(y)

plt.show()
```

```python
# iske andar hm label bhi show kra skte hai
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x, labels=["Python", "C", "C++", "Java", "JavaScript"])
plt.yticks(y)

plt.show()
```

```python
# we can also increase the limit of x and y(axis)
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xlim(0,10) # for the x limit 
plt.ylim(0,10) # for the y limit

plt.show()
```

```python
# hm axis ke andar limit decide kr skte hai ki hamara data kha se kha tk show hoga inside the x and y axis.

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.axis([0,10,0,7]) # for the x(0,10) and y(0,7) 

plt.show()
```

# Text in Matplotlib


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("WsCube",fontsize=15) # we can also give the font size inside this
# hm x and y ye bhi data ko bta skte hai ki vo kya represent kr rha hai
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.show()
```

```python
# agar hme graph ke upar kuchh bhi show krna hai then use case me hm text ka use karte hai
# and hme text lagane se phle uski position bhi decide kr skte hai ki vo kha pr aayega. text(x,y,"text")

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("WsCube",fontsize=15) 
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"java")

plt.show()
```

```python
# now is java pe hm operation perform kr skte hai like fontsize,style,bbox(to make the text inside the box with the given color)
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("WsCube",fontsize=15) 
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"java",fontsize=15,style="italic",bbox={"facecolor":"red"})

plt.show()
```

```python
# hm annotate() function ke through kisi bhi point pr text daal skte hai and uske andar arrow bhi display kra skte hai

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("WsCube",fontsize=15) 
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"java",fontsize=15,style="italic",bbox={"facecolor":"red"})

plt.annotate("python",xy=(2, 1),xytext=(4, 4), arrowprops=dict(facecolor="black", shrink=0.05))
# xy() isme hm arrow ke position ko decide krte hai
# xytext() isme hm text ka position ko decide krte hai

# hm legend ka use karke line kis bare me show kr rhi hai uske text ko bhi decide kr skte hai
plt.legend(["up"],loc=5,facecolor="Red",edgecolor="black",framealpha=0.5,shadow=True) # isme position left side se start hota hai and middle me bhi loc count hota hai(total=10 loc)
# agar isme hme light dim rkhana hai then use framealpha
plt.show()
```
